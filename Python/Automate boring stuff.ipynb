```python
def printTable(table):
    colWidths = [0] * len(table)
    c = 0
    while c < len(table):
        for items in table[c]:
            if colWidths[c] < len(items):
                colWidths[c] = len(items)
        c = c + 1
        
    for item in range(len(table[0])):
        for lst in range(len(table)):
            print(table[lst][item].rjust(colWidths[lst]," "), end=' ')
        print()  
            
        
```

```python
def print_table(table):
    """Print a formatted, properly aligned table version of a list."""
    # Create list with zeroes equal to the length of input list
    col_widths = [0] * len(table)

    # Finds longest word in each sublists and sets col_width[x] to value
    count = 0
    while count < len(table):
        for item in table[count]:
            if len(item) > col_widths[count]:
                col_widths[count] = len(item)
        count += 1

    # Iterates over lists taking nth element from each
    # and printing it with the other nth subitems
    # right aligned according to the corresponding col_width value

    for word in range(len(table[0])):
        for item in range(len(table)):
            print(table[item][word].rjust(col_widths[item]), end=' ')
        print()
        
    print(col_widths)    


print_table(tableData)
```

```python
def isPhoneNumber(text):
    if len(text) != 12:
        return False
    for i in range(0,3):
        if not text[i].isdecimal():
            return False
    if text[3] != '-':
        return False
    for i in range(4,7):
        if not text[i].isdecimal():
            return False
    if text[7] != '-':
        return False
    for i in range(8,12):
        if not text[i].isdecimal():
            return False
    return True    
```

```python
print('415-555-4242 is a phone number:')
print(isPhoneNumber('415-555-4242'))
print('Moshi moshi is a phone number:')
print(isPhoneNumber('Moshi moshi'))
```

### Greedy and Nongreedy Matching

Regex Version of strip()

```python
def strip(text, remove=''):
    """Perform string.strip-like functions but using regexes."""
    # Removes whitespace from either end of the string
    if remove == '':
        space_regex = re.compile(r'^(\s*)(.*)(\s)*$') #Starting from zero or more space, zero or more characters, space, 
        trimmed = space_regex.search(text)            #zero or more of total word/sentence, end.
        return trimmed.group(2) # returns (.*) which means what was in that word or sentence.

    # Removes character inputted as 2nd argument from ends of string
    else:
        remove_start = re.compile(r'^([%s]+)' % remove) #Start, one or more space
        remove_end = re.compile(r'([%s]+)$' % remove)
        start = remove_start.search(text)
        end = remove_end.search(text)
        # Allows function to strip even if only one side has remove characters
        try:
            return text[len(start.group()):len(text) - len(end.group())]
        except AttributeError:
            error_avoid = remove + text + remove
            return strip(error_avoid, remove)


# Get function arguments from user then print stripped string
user_text = input('Enter the text you would like stripped here: ')
user_remove = input('Enter the character you want stripped here'
                    ' (Removes Space as Default): ')

print(strip(user_text, user_remove))
```

### Raising Exceptions

```python
raise Exception('This is the error message.')
```

```python
def boxPrint(symbol, width, height):
    if len(symbol) != 1:
        raise Exception('Symbol must be a single character string.')
    if width <= 2:
        raise Exception('Width must be greater than 2.')
    if height <= 2:
        raise Exception('Height must be greater than 2.')
    print(symbol * width)                            # in case of 4x4: prints **** (* times 4) 
    for i in range(height - 2):                      # for i in 1:2, because it's not printing top or bottom of the box
        print(symbol + (' ' * (width - 2)) + symbol) # prints *  * two times
    print(symbol * width)                            # prints ****
    
for sym, w, h in (('*', 4, 4), ('O', 20, 5), ('x', 1, 3), ('ZZ', 3, 3)): # exeptions because, width is 1 and ZZ in not 1 symbol
    try:
        boxPrint(sym, w, h)
    except Exception as err:
        print('An exception happened: ' + str(err))
```
